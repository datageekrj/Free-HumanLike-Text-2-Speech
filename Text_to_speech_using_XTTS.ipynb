```python
def create_sh_script(txt_file_name, speaker_id = "Viktor Eka", words_to_process_once = 10):
    # Name of the shell script to be created
    sh_file_name = "run_tts.sh"

    # Content of the shell script
    sh_content = f"""#!/bin/bash

# Path to the file containing your text
TEXT_FILE="{txt_file_name}"

# Read the entire file and extract all words
all_words=($(cat "$TEXT_FILE"))

# Define the number of words to process at a time
words_per_chunk={words_to_process_once}

# Initialize a counter for chunk processing
counter=0

# Array to hold the names of all generated wav files
output_files=()

# Loop through the array of words
while [ ${{#all_words[@]}} -gt 0 ]
do
    # Take the first 10 words
    word_chunk="${{all_words[@]:0:$words_per_chunk}}"

    # Join the 10 words into a single string (part1)
    part1=$(echo "$word_chunk" | tr ' ' ' ')

    # Define output file name
    output_file="out_chunk_$counter.wav"

    # Construct and run the tts command
    tts --text "$part1" --model_name "tts_models/multilingual/multi-dataset/xtts_v2" --out_path "$output_file" --speaker_idx "{speaker_id}" --language_idx "en"

    # Add the output file to the list
    output_files+=("$output_file")

    # Remove the processed 10 words from the array
    all_words=("${{all_words[@]:$words_per_chunk}}")

    # Increment counter for naming output files
    counter=$((counter+1))

done

# Use sox to concatenate all the generated wav files into one consolidated file
sox "${{output_files[@]}}" consolidated_output.wav

# Inform the user
echo "All wav files have been combined into consolidated_output.wav"
"""

    # Write the shell script content to a file
    with open(sh_file_name, 'w') as sh_file:
        sh_file.write(sh_content)

    # Give execution permission to the shell script
    import os
    os.chmod(sh_file_name, 0o755)

    print(f"{sh_file_name} has been created and is ready to run.")
```
